# 通过CNN实现手写汉字数字识别(应用)

这里实现了用测试集数据检验训练出来的模型效果如何

```python
import tensorflow as tf
import numpy as np
from model_cnn import MyCNN
import os
from matplotlib import pyplot as plt
from creat_dataset import get_test_data
print(tf.__version__)
```

载入测试集数据

```python
x_test, y_test = get_test_data()
x_test = x_test.astype(np.float32)
y_test = y_test.astype(np.float32)
#打乱数据
np.random.seed(96)
np.random.shuffle(x_test)
np.random.seed(96)
np.random.shuffle(y_test)
```

导入训练好的模型

```python
model = MyCNN()
checkpoint_save_path = "./checkpoint/mycnn.ckpt"
model.load_weights(checkpoint_save_path)
```

将测试集数据导入训练模型，计算准确率

```python
predict = model.predict(x_test, batch_size=100)
y_pred = tf.argmax(predict, axis=1)
correct = tf.where(y_pred == y_test, 1, 0)
print("accuracy:", np.sum(correct) / len(correct))
```
